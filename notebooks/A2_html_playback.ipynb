# A2 HTML playback and visualization

This notebook is the browser side of the simulation workflow.
The simulation stays in Python. HTML and JavaScript are used for playback and visualization.


## Goal

- Export an interactive travel graph explorer.
- Keep the browser layer separate from the simulation engine.
- Prepare a clean handoff for future playback frames.


```python
from pathlib import Path
from IPython.display import IFrame, display

from utils import TravelGraphManager, TravelGraphVisualizer, SimulationConfig

def find_repo_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / "configs" / "travel_graph_config.yaml").exists():
            return candidate
    raise FileNotFoundError("Could not find repo root from the current notebook location.")

ROOT = find_repo_root(Path.cwd().resolve())
cfg = SimulationConfig.from_yaml(ROOT / "configs" / "travel_graph_config.yaml")
ROOT


config_summary = {
    "walk_wt": cfg.walk_wt,
    "ride_wt": cfg.ride_wt,
    "wait_wt": cfg.wait_wt,
    "transfer_wt": cfg.transfer_wt,
    "v_jeep": cfg.v_jeep,
    "v_passenger": cfg.v_passenger,
    "headway_s": cfg.headway_s,
    "passenger_generation_interval_dt": cfg.passenger_generation_interval_dt,
    "passenger_generation_mean": cfg.passenger_generation_mean,
    "passenger_generation_std": cfg.passenger_generation_std,
}
config_summary
```

```python
mgr = TravelGraphManager(
    edges_csv=ROOT / "data" / "iligan_travel_graph.csv",
    nodes_csv=ROOT / "data" / "travel_graph_nodes.csv",
    walk_wt=cfg.walk_wt,
    ride_wt=cfg.ride_wt,
    wait_wt=cfg.wait_wt,
    transfer_wt=cfg.transfer_wt,
)

visualizer = TravelGraphVisualizer(mgr, title="A2 HTML Playback Explorer")
html_path = visualizer.export(ROOT / "results" / "A2_html_playback" / "interactive_travel_graph_explorer.html")
display(IFrame(str(html_path), width=1200, height=800))
html_path
```

## Notes

This notebook only prepares the browser layer.
Future playback can consume exported frames or snapshots without moving the simulation into HTML.
